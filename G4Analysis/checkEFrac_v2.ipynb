```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cosmicFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/cosmicEFrac.csv'
signalFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/nueArCCSignalEFrac.csv'
nueOutFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/nueArCCoutEFrac.csv'
brnFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/brnEFrac.csv'
hogFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/hogEFrac.csv'
```

```python
cosmic = pd.read_csv( cosmicFile )
signal = pd.read_csv( signalFile )
nueOut = pd.read_csv( nueOutFile )
brn = pd.read_csv( brnFile )
hog = pd.read_csv( hogFile )
```

```python
cosmic['SignalVolE'] = cosmic.FiducialE + cosmic.PartConE
signal['SignalVolE'] = signal.FiducialE + signal.PartConE
nueOut['SignalVolE'] = nueOut.FiducialE + nueOut.PartConE
brn['SignalVolE'] = brn.FiducialE + brn.PartConE
hog['SignalVolE'] = hog.FiducialE + hog.PartConE
```

```python
fig, ax = plt.subplots(1, 5, figsize = (20, 3))

ax[0].hist( signal.SignalVolE, bins = np.linspace(0, 40, 41), histtype = 'step', lw = 2, label = r'$\nu_e$-Ar CC Signal')
ax[1].hist( cosmic.SignalVolE, bins = np.linspace(0, 40, 41), histtype = 'step', lw = 2, label = 'Cosmic')
ax[2].hist( brn.SignalVolE, bins = np.linspace(0, 40, 41), histtype = 'step', lw = 2, label = 'BRN')
ax[3].hist( nueOut.SignalVolE, bins = np.linspace(0, 40, 41), histtype = 'step', lw = 2, label = r'$\nu_e$-Ar CC Outside')
ax[4].hist( hog.SignalVolE, bins = np.linspace(0, 40, 41), histtype = 'step', lw = 2, label = 'Hot-off Gas')



for iax in ax:
    iax.set_xlabel('Energy Deposited in the Signal Volume')
    iax.set_xlim([0, 40])
    iax.grid(':')
    iax.legend()
```
