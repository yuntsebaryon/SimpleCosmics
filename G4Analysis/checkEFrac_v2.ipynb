# Energy Deposition in LArTPC

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cosmicFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/cosmicEFrac.csv'
signalFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/nueArCCSignalEFrac.csv'
nueOutFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/nueArCCoutFiducialEFrac.csv'
```

```python
cosmic = pd.read_csv( cosmicFile )
signal = pd.read_csv( signalFile )
nueOut = pd.read_csv( nueOutFile )
```

```python
cosmic
```

```python
cosmic['SignalVolEFrac'] = ( cosmic.FiducialE + cosmic.PartConE )/ cosmic.TotalE
cosmic
```

```python
signal['SignalVolEFrac'] = ( signal.FiducialE + signal.PartConE )/ signal.TotalE
nueOut['SignalVolEFrac'] = ( nueOut.FiducialE + nueOut.PartConE )/ nueOut.TotalE
```

```python
fig, ax = plt.subplots(1, 3, figsize = (12, 3))

ax[0].hist( signal.SignalVolEFrac, bins = np.linspace(0, 1, 41), histtype = 'step', lw = 2, label = r'$\nu_e$-Ar CC Signal')
ax[1].hist( cosmic.SignalVolEFrac, bins = np.linspace(0, 1, 41), histtype = 'step', lw = 2, label = 'Cosmic')
ax[2].hist( nueOut.SignalVolEFrac, bins = np.linspace(0, 1, 41), histtype = 'step', lw = 2, label = r'$\nu_e$-Ar CC Outside')

for iax in ax:
    iax.set_xlabel('Energy Fraction in the Signal Volume')
    iax.set_xlim([0, 1])
    iax.grid(':')
    iax.legend( loc = 'upper center')
```

```python
fig, ax = plt.subplots( figsize = (8, 6) )

ax.hist( signal.SignalVolEFrac, bins = np.linspace(0, 1, 41), density = True, histtype = 'step', lw = 2, 
        label = r'$\nu_e$-Ar CC Signal' )
ax.hist( cosmic.SignalVolEFrac, bins = np.linspace(0, 1, 41), density = True, histtype = 'step', lw = 2,
        label = 'Cosmic' )
ax.hist( nueOut.SignalVolEFrac, bins = np.linspace(0, 1, 41), density = True, histtype = 'step', lw = 2, linestyle = '-.',
        label = r'$\nu_e$-Ar CC Outside' )

ax.set_xlabel( 'Deposited Energy Fraction in the Signal Volume', fontsize = 16)
ax.set_ylabel( 'Arbitray Unit', fontsize = 16)
ax.tick_params( axis = 'x', labelsize = 16)
ax.tick_params( axis = 'y', labelsize = 16)

ax.set_xlim([ 0, 1 ])
ax.grid(':')
ax.legend(fontsize = 16)

# fig.savefig('EFraction.pdf')
```

### Scales of each samples

|        |  Expected number of events per year | Generated number of events | Scale |
|--------|-------------------------------------|----------------------------|-------|
| Cosmics | 1.08e9 | 2.5e6 | 432 |
| $\nu_e$-Ar CC signal | 116.5 | 1e5 | 116.5/1e5 |
| $\nu_e$-Ar CC out fiducial | 813 | 4e5 | 813/4e5 |

```python
fCosmic = 432.
fNueSignal = 116.5/1e5
fNueOutFiducial = 813./4e5
fNueSignal, fNueOutFiducial
```

```python
cosmicCut = cosmic[(cosmic.SignalVolEFrac>0.4)&(cosmic.FiducialEFrac>0.025)&(cosmic.MaxE>5.)&(cosmic.TotalE<52.)&(cosmic.OutE>0.1)]
signalCut = signal[(signal.SignalVolEFrac>0.4)&(signal.FiducialEFrac>0.025)&(signal.MaxE>5.)&(signal.TotalE<52.)&(signal.OutE>0.1)]
nueOutCut = nueOut[(nueOut.SignalVolEFrac>0.4)&(nueOut.FiducialEFrac>0.025)&(nueOut.MaxE>5.)&(nueOut.TotalE<52.)&(nueOut.OutE>0.1)]
```

```python
SF = 100.
fig, ax = plt.subplots( figsize = (8, 6), layout = 'tight' )

ax.hist(cosmicCut.MaxE, weights = np.full_like(cosmicCut.MaxE, fCosmic), bins = np.linspace(0, 40, 41),
        histtype = 'step', lw = 2, label = 'Cosmic with no muon>5cm')
ax.hist(signalCut.MaxE, weights = np.full_like(signalCut.MaxE, fNueSignal*SF), bins = np.linspace(0, 40, 41),
        histtype = 'step', lw = 2, label = r'$\nu_e$-Ar CC signal')
ax.hist(nueOutCut.MaxE, weights = np.full_like(nueOutCut.MaxE, fNueOutFiducial*SF), bins = np.linspace(0, 40, 41), 
        histtype = 'step', lw = 2, label = r'$\nu_e$-Ar CC outside')
ax.set_xlabel( r'Leading Energy [MeV]', fontsize = 16 )
ax.set_ylabel( r'Events/MeV/Year', fontsize = 16 )
ax.tick_params(axis = 'x', labelsize = 16)
ax.tick_params(axis = 'y', labelsize = 16)
ax.set_xlim([5, 40])
#ax.set_ylim( [0, 30000])
ax.legend( fontsize = 16)

ax.grid(":")
```
