```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
signalFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/nueArCCSignalEFrac.csv'
hogFile = '/Users/yuntse/data/coherent/preLArTPC/analysis/EFracv2/hogEFrac.csv'
```

```python
signal = pd.read_csv( signalFile )
hog = pd.read_csv( hogFile )
```

```python
hog
```

```python
fig, ax = plt.subplots(1, 2, figsize = (8, 3))

ax[0].hist( signal.MaxE, bins = np.linspace(0, 60, 61), histtype = 'step', lw = 2, label = r'$\nu_e$-Ar CC Signal')
ax[1].hist( hog.MaxE, bins = np.linspace(0, 1, 61), histtype = 'step', lw = 2, label = 'HOG')
ax[0].set_xlim([0, 60])
ax[1].set_xlim([0, 1])

for iax in ax:
    iax.set_xlabel('Leading Charged Cluster Energy [MeV]')
    iax.grid(':')
    iax.legend()
```
